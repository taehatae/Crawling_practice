```python
import pandas as pd

# 1. 파일 불러오기
# 한글 깨짐 방지를 위해 encoding='utf-8-sig'를 주로 사용합니다. 
# 만약 에러가 난다면 'cp949'로 변경해 보세요.
file_path = 'Hormuz_Strait_Final_Cleaned.csv'
df_total = pd.read_csv(file_path, encoding='utf-8-sig')

# 2. 데이터 구조 확인
print("✅ 데이터 로드 완료!")
print(f"전체 데이터 행 수: {len(df_total)}")

# 3. 컬럼 구성 확인 
# 감성 분석에 사용할 'doc_final' 컬럼이 잘 있는지 확인합니다.
display(df_total.head())

# 4. 결측치 확인 및 제거 (필요 시)
# 텍스트 데이터가 비어있으면 감성 분석 시 에러가 날 수 있으므로 빈 값을 체크합니다.
df_total = df_total.dropna(subset=['doc_final'])
print(f"결측치 제거 후 행 수: {len(df_total)}")
```

```python
import json
import pandas as pd
from tqdm import tqdm

# 1. KNU 감성사전 불러오기
# 역슬래시(\)로 인한 경로 오류를 방지하기 위해 경로 앞에 r을 붙여줍니다.
file_path = r'C:\Users\황태하\Crawling_practice\SentiWord_info.json'

with open(file_path, 'r', encoding='utf-8-sig') as f:
    senti_dict_list = json.load(f)

# 2. 검색 효율을 위해 리스트 형태의 사전을 {단어: 점수} 딕셔너리로 변환
# 사전의 키값(word, polarity)은 실제 파일 구조에 맞게 조정이 필요할 수 있습니다.
senti_map = {item['word']: int(item['polarity']) for item in senti_dict_list}

# 3. 문서별 감성 점수 계산 함수 정의
def get_sentiment_score(text):
    if not isinstance(text, str):
        return 0
    
    score = 0
    words = text.split()  # 형태소 분석된 결과인 doc_final 활용
    count = 0
    
    for word in words:
        if word in senti_map:
            score += senti_map[word]
            count += 1
            
    # [참고] 단어가 하나도 매칭되지 않으면 0점, 매칭되면 평균 점수를 반환하거나 합산 점수 사용
    # 여기서는 직관적인 합산 점수를 사용하되, 문서 길이에 따른 편차를 줄이려면 평균을 사용해도 좋습니다.
    return score

# 4. 데이터프레임에 감성 점수 및 라벨 삽입
print("📊 감성 분석 진행 중...")
tqdm.pandas()  # 진행률 표시를 위한 설정

# 감성 점수 계산
df_total['sentiment_score'] = df_total['doc_final'].progress_apply(get_sentiment_score)

# 점수에 따른 긍정/부정/중립 라벨링
def label_sentiment(score):
    if score > 0: return 'Positive'
    elif score < 0: return 'Negative'
    else: return 'Neutral'

df_total['sentiment_result'] = df_total['sentiment_score'].apply(label_sentiment)

# 5. 결과 확인
print("\n✅ 감성 분석 완료!")
display(df_total[['doc_final', 'sentiment_score', 'sentiment_result']].head())
```

### LDA 분석 먼저 시행

```python
import gensim
from gensim import corpora
from gensim.models import LdaModel
import pandas as pd

# 1. 텍스트 토큰화 및 전처리 확인
# df_total['doc_final']의 텍스트를 공백 기준으로 분리하여 리스트 형태로 변환
tokenized_docs = df_total['doc_final'].apply(lambda x: str(x).split())

# 2. 단어 사전(Dictionary) 생성 및 필터링
dictionary = corpora.Dictionary(tokenized_docs)

# 지나치게 빈도가 낮거나 높은 단어를 제거하여 토픽의 변별력 강화
# no_below=5: 5번 미만 등장 단어 제거 / no_above=0.5: 전체 문서의 50% 이상 등장 단어 제거
dictionary.filter_extremes(no_below=5, no_above=0.5)

# 3. 코퍼스(Corpus) 생성: 문서를 Bag-of-Words(ID, 빈도) 형태로 변환
corpus = [dictionary.doc2bow(doc) for doc in tokenized_docs]

# 4. LDA 모델 학습 (K-Means 결과와 동일하게 num_topics=5 설정)
num_topics = 5 
lda_model = LdaModel(
    corpus=corpus, 
    id2word=dictionary, 
    num_topics=num_topics, 
    passes=20,          # 학습 횟수를 늘려 토픽 응집도 향상
    iterations=400,     # 각 문서당 반복 횟수 최적화
    random_state=42     # 결과 재현성을 위한 난수 고정
)

# 5. 토픽별 핵심 단어 출력
print("✨ LDA 토픽 모델링 결과 (K=5):")
for i, topic in lda_model.print_topics(num_words=10):
    # 단어와 가중치를 깔끔하게 분리하여 출력
    print(f"🔹 Topic {i}: {topic}")
    print("-" * 100)

# 6. 모든 문서의 주력 토픽(Dominant Topic)을 데이터프레임에 할당
def get_main_topic(bow):
    # 해당 문서에서 가장 확률이 높은 토픽 번호 반환
    topic_probs = lda_model.get_document_topics(bow)
    return sorted(topic_probs, key=lambda x: x[1], reverse=True)[0][0]

# 각 문서의 코퍼스를 기반으로 가장 비중이 높은 토픽 번호를 추출하여 새 컬럼 생성
df_total['lda_topic'] = [get_main_topic(c) for c in corpus]

print("\n✅ LDA 분석 및 문서별 토픽 할당 완료!")
# 첫 번째 문서의 분석 결과 예시 출력
print(f"📝 첫 번째 문서의 주력 토픽 번호: {df_total['lda_topic'].iloc[0]}")
print(f"📊 첫 번째 문서의 상세 토픽 비중: {lda_model.get_document_topics(corpus[0])}")
```

```python
# 1. 모델에서 토픽별 단어 추출 (K=5에 맞춰 num_topics=5 설정)
# formatted=False로 설정해야 단어와 가중치(확률)를 튜플 형태로 가져오는게 가능!
topics = lda_model.show_topics(num_topics=5, num_words=10, formatted=False)

topic_data = []

# 2. 리스트 컴프리헨션을 활용하여 데이터 정리
for i, topic in topics:
    # 각 토픽의 단어만 추출하여 리스트 생성
    words = [word for word, prob in topic]
    # 'Topic 번호'를 첫 번째 요소로 넣고 뒤에 단어 리스트를 합침
    topic_data.append([f"Topic {i}"] + words)

# 3. 데이터프레임 생성
# 컬럼명은 'Topic'과 '단어 1' ~ '단어 10'으로 구성
column_names = ['Topic'] + [f'Word {j+1}' for j in range(10)]
lda_df = pd.DataFrame(topic_data, columns=column_names)

# 4. 가독성을 위해 행과 열을 바꾼(Transpose) 버전 생성
# 토픽이 많아질수록 가로로 긴 것보다 세로로 긴 형태가 비교하기 편합니다.
lda_df_t = lda_df.set_index('Topic').T

print("✨ LDA 토픽별 핵심 단어 요약 (세로형 보기):")
display(lda_df_t)
```

### 토픽별 감성 비율 및 빈도

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# 한글 폰트 설정
plt.rcParams['font.family'] = 'Malgun Gothic'

# 1. 감성 클래스별 비율 (Pie Chart)
sentiment_counts = df_total['sentiment_result'].value_counts()
plt.figure(figsize=(7, 7))
plt.pie(sentiment_counts, labels=sentiment_counts.index, autopct='%1.1f%%', startangle=140, colors=['#ff9999', '#e0e0e0', '#66b3ff'])
plt.title('전체 감성 클래스 비율')
plt.show()

# 2. 토픽별 감성 빈도/비율 데이터프레임 생성 및 시각화
topic_sentiment_df = pd.crosstab(df_total['lda_topic'], df_total['sentiment_result'])
topic_sentiment_pct = topic_sentiment_df.div(topic_sentiment_df.sum(axis=1), axis=0)

# 시각화 (Stacked Bar)
topic_sentiment_pct.plot(kind='bar', stacked=True, figsize=(10, 6), color=['#ff9999', '#e0e0e0', '#66b3ff'])
plt.title('토픽별 감성 분포 비교')
plt.legend(title='감성', bbox_to_anchor=(1.05, 1))
plt.show()
```

### 본문 vs 댓글 감성 분석 시행

```python
# [이미 확인된 컬럼명 활용]
# doc_final: 블로그/카페 포스팅 본문
# comment_final: 해당 포스팅의 댓글들

# 1. 본문 데이터 감성 분석 (작성자의 시각)
df_article = df_total[['doc_final']].copy().rename(columns={'doc_final': 'text'})
df_article['source'] = 'Post_Body' # 'Article' 대신 'Post_Body'로 변경
df_article['sentiment_score'] = df_article['text'].apply(get_sentiment_score)
df_article['sentiment_result'] = df_article['sentiment_score'].apply(label_sentiment)

# 2. 댓글 데이터 감성 분석 (독자의 반응)
df_comment = df_total[df_total['comment_final'].notnull()].copy()[['comment_final']]
df_comment = df_comment.rename(columns={'comment_final': 'text'})
df_comment['source'] = 'Post_Comment' # 'Comment' 대신 'Post_Comment'로 변경
df_comment['sentiment_score'] = df_comment['text'].apply(get_sentiment_score)
df_comment['sentiment_result'] = df_comment['sentiment_score'].apply(label_sentiment)
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd

# 한글 폰트 설정 (윈도우 기준)
plt.rcParams['font.family'] = 'Malgun Gothic'
plt.rcParams['axes.unicode_minus'] = False

# 1. 시각화를 위한 데이터 요약 (비중 계산)
# 정규화(normalize=True)를 통해 데이터 개수 차이와 상관없이 '비율'로 비교합니다.
body_res = df_article['sentiment_result'].value_counts(normalize=True).sort_index()
comment_res = df_comment['sentiment_result'].value_counts(normalize=True).sort_index()

# 2. 비교 데이터프레임 생성
comparison_df = pd.DataFrame({
    '블로그/카페 본문 (Post)': body_res,
    '사용자 댓글 (Comment)': comment_res
}).fillna(0).T

# 3. 시각화 (Grouped Bar Chart)
# 색상 설정: 부정(Coral), 중립(Silver), 긍정(Skyblue)
colors = ['#ff9999', '#e0e0e0', '#66b3ff']

ax = comparison_df.plot(kind='bar', figsize=(13, 8), color=colors, width=0.8)

# 그래프 꾸미기
plt.title('호르무즈 위기 관련 블로그/카페 [본문 vs 댓글] 감성 비교', fontsize=20, fontweight='bold', pad=30)
plt.ylabel('비중 (Ratio)', fontsize=14)
plt.xlabel('데이터 소스 (Data Source)', fontsize=14)
plt.xticks(rotation=0, fontsize=12)
plt.legend(title='감성 결과', bbox_to_anchor=(1.05, 1), loc='upper left', fontsize=12)
plt.grid(axis='y', linestyle='--', alpha=0.3)

# 막대 위에 수치(%) 표시
for p in ax.patches:
    height = p.get_height()
    if height > 0: # 0보다 큰 경우에만 표시
        ax.annotate(f'{height:.1%}', 
                    (p.get_x() + p.get_width() / 2., height + 0.01), 
                    ha='center', va='bottom', fontsize=12, fontweight='bold')

plt.tight_layout()
plt.show()
```
